```python
# prompt: merge the target and matches using inner join  wrt to match id and match id should appear only one
result = pd.merge(targets, matches, left_on='match_id', right_on='id')
result = result.drop('id', axis=1)
result
```

```python
# prompt: create 2 new coloumns batting team and bowling team such that if win_by_runs is 0 then batting team will be the winner and bowling team will be the other one and if win_by _runs is positive  then batting team will be losing team and bowling team will be winning team

def assign_teams(row):
  if row['win_by_runs'] == 0:
    batting_team = row['winner']
    bowling_team = row['team1'] if row['winner'] == row['team2'] else row['team2']
  elif row['win_by_runs'] > 0:
    batting_team = row['team2'] if row['winner'] == row['team1'] else row['team1']
    bowling_team = row['winner']
  else:
    batting_team = None
    bowling_team = None
  return batting_team, bowling_team

result[['batting_team', 'bowling_team']] = result.apply(assign_teams, axis=1, result_type='expand')
result
```

```python
X=final_result.drop('result',axis=1)
y=final_result['result']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','venue'])
],remainder='passthrough')
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
pipe.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=pipe.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
pipe2=Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])
pipe2.fit(X_train,y_train)
y_pred=pipe2.predict(X_test)
accuracy_score(y_test,y_pred)
```
